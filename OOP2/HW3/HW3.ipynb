```python
# Q1
import random

class Player(object):
    def __init__(self, name = 'player', level = 0, ch_type = 'default') -> None:
        self.name = name
        self.level = level
        self.character_type = ch_type
    
    def __str__(self) -> str:
        return f"Name: {self.name}\nLevel: {self.level}\nCharacter Type: {self.character_type}"

    def __repr__(self):
        return self.__str__()      

    def q1(self, other):
        '''
        function required by q1 print the player with higher level
        only compare same type
        '''
        if self.level > other.level:
            print(self)
        elif self.level < other.level:
            print(other)
        else:
            print('tie')

player1 = Player('player1', random.randint(1,99))
player2 = Player('player2', random.randint(1,99))
player1.q1(player2)
```

```python
# Q2
import random

class RandomSeatingPlan(object):
    def __init__(self, n = 10, m = 30) -> None:
        self.seating_plan = []
        if n == 1:
            print('Only one studednt here so sitting anywhere is fine')
            return
        
        if n > m:
            print('There are more students than seats available')
            return

        spaces = m - n
        space = spaces // (n - 1)
        resid = spaces % (n - 1)

        students = list(range(n))
        random.shuffle(students)
        plus_one = students[:resid]

        emt = '_'
        count = 0
        while students: #if use for loop can use i instead of count
            curr = students.pop()
            if students:
                if count in plus_one:
                    self.seating_plan.append(curr)
                    self.seating_plan += emt * (space + 1)
                else:
                    self.seating_plan.append(curr)
                    self.seating_plan += emt * space
            else: #last student
                if count in plus_one:
                    self.seating_plan += emt
                    self.seating_plan.append(curr)
                else:
                    self.seating_plan.append(curr)
            count += 1        

    def findSeat(self, student_id = 0):
        try:
            idx = self.seating_plan.index(student_id)
        except ValueError:
            print('student not found')
            idx = -1
        return idx

# generate #of students and #of seats n <= m
n_max = 50
n = random.randint(1, n_max)
m = random.randint(n, n * 5)

# n = 3 #test case
# m = 11

q2 = RandomSeatingPlan(n, m)
print(q2.seating_plan)
print(q2.findSeat(1))
```

```python
# Q3
#may use os module for path
class TextDocument(object):
    def __init__(self, path) -> None:
        self.content = ''
        try:
            with open(path, 'r') as fileObj:
                self.content = fileObj.read()
        except IOError:
            print('Unable to open file, please try again')
    
    def __add__(self, other):
        try:
            with open('newFile.txt', 'w') as out:
                try:
                    out.write(self.content)
                    out.write(other.content)
                except IOError:
                    print('Cannot write into the output file')
        except(PermissionError, OSError):
            print('Cannot open the output file')

#test case
td1 = TextDocument('TextFile1.txt')
td2 = TextDocument('TextFile2.txt')
td1+td2
```

```python
# from decimal import Decimal
class Color(object):
    def __init__(self, r = 0, g = 0, b = 0) -> None:
        for i in [r, g, b]:
            if i < 0 or i > 1:
                print('RGB value should lie between 0 and 1 (inclusive)')
                print('object will be initialized to (0,0,0)')
                r, g, b = 0, 0, 0
                break
        self.r = r
        self.g = g
        self.b = b

    def __add__(self, other):
        r = self.r + other.r
        g = self.g + other.g
        b = self.b + other.b
        r, g, b = self.satuation(r, g, b)
        return Color(r, g, b)

    def __sub__(self, other):
        r = self.r - other.r
        g = self.g - other.g
        b = self.b - other.b
        r, g, b = self.satuation(r, g, b)
        return Color(r, g, b)

    def __str__(self):
        return f"(r:{self.r}, g:{self.g}, b:{self.b})"

    def __repr__(self):
        return self.__str__()
        
    def satuation(self, r, g, b):
        ls = [r,g,b]
        for i in range(3):
            if ls[i] < 0:
                ls[i] = 0
            elif ls[i] > 1:
                ls[i] = 1
        return ls

a = Color(1, 0.2, 0.45)
b = Color(1, 0.3, 0.8)
print(a)
print(b)
c = a + b
d = a - b
print(a)
print(b)
print(c)
print(d)
```

```python
# Q5
import math

class Logarithm(object):
    def __init__(self, base = 2, number = 1) -> None:
        if base <= 0: #negative base can be valid if applied to complex number but its not considered here
            print(f'Base entered should be positive, {base} is invalid, and is initialized to 10')
            base = 2
        if number <= 0:
            print(f'Number entered should be positive, {number} is invalid, and is initialized to 1')
            number = 1
            
        self.base = base
        self.number = number
    
    def __str__(self):
        return f'logrithm(base:{self.base}, num:{self.number})'
    
    def __repr__(self):
        return self.__str__()

    def __add__(self, other):
        if type(other) == int or type(other) == float:
            return math.log(self.number, self.base) + other

        elif type(other) == Logarithm:
            if self.base == other.base:
                number = self.number * other.number
                return Logarithm(self.base, number)
            else: #different bases
                base = 2 #change the base to 2
                left = math.log(self.number, self.base)
                right = math.log(other.number, other.base)
                return Logarithm(base, base ** (left + right))
        else:
            print('unsupported operation')
            raise TypeError

    def __sub__(self, other):
        if type(other) == int or type(other) == float:
            return math.log(self.number, self.base) - other

        elif type(other) == Logarithm:
            if self.base == other.base:
                number = self.number / other.number
                return Logarithm(self.base, number)
            else: #different bases
                base = 2
                left = math.log(self.number, self.base)
                right = math.log(other.number, other.base)
                return Logarithm(base, base ** (left - right))
        else:
            print('unsupported operation')
            raise TypeError

    # try also to support left operand with int and float
    def __radd__(self, other):
        return  self.__add__(other)

    def __rsub__(self, other):
        return  -self.__sub__(other)


a = Logarithm(2, 16)
b = Logarithm(8, 64)
c = 5 + a
d = 5 - b
print(a, b)
print(c, d)
```

```python
# Q6
def gcd(bigger, smaller):
    if smaller > bigger:
        bigger, smaller = smaller, bigger
    while smaller != 0:
        remainder = bigger % smaller
        bigger, smaller = smaller, remainder
    return bigger

def lcm(a, b):
    return (a * b) // gcd(a, b)

class Rational(object):
    # I skipped the print('in sth') part of the code
    def __init__ (self, numer, denom=1):
        self.numer = numer
        self.denom = denom

    def __str__(self):
        return f'{self.numer}/{self.denom}'

    def __repr__(self):
        return self.__str__()

    def __add__(self, other):
        if type(other) == int:
            other = Rational(other)

        if type(other) == Rational:
            newDenom = lcm(self.denom, other.denom)
            newNumer = self.numer/self.denom * newDenom + other.numer/other.denom * newDenom
            return Rational(newNumer, newDenom)
        else:
            print('unsupported operation')
            raise TypeError

    def __sub__(self, other):
        if type(other) == int:
            other = Rational(other)

        if type(other) == Rational:
            newDenom = lcm(self.denom, other.denom)
            newNumer = self.numer/self.denom * newDenom - other.numer/other.denom * newDenom
            return Rational(newNumer, newDenom)
        else:
            print('unsupported operation')
            raise TypeError

    def __mul__(self, other):
        if type(other) == int:
            other = Rational(other)

        if type(other) == Rational:
            newDenom = self.denom * other.denom
            newNumer = self.numer * other.numer
            return Rational(newNumer, newDenom)
        else:
            print('unsupported operation')
            raise TypeError

    def __truediv__(self ,other):
        if type(other) == int:
            other = Rational(other)

        if type(other) == Rational:
            inverted_other = Rational(other.denom, other.numer)
            return self.__mul__(inverted_other)
        else:
            print('unsupported operation')
            raise TypeError

    # should also support left operand with int
    def __radd__(self, other):
        return  self.__add__(other)

    def __rsub__(self, other):
        temp = self.__sub__(other)
        return  Rational(-temp.numer, temp.denom)

    def __rmul__(self, other):
        return self.__mul__(other)
    
    def __rtruediv__(self, other):
        inverted_self = Rational(self.denom, self.numer)
        return other * inverted_self

    def reduce_rational(self):
        the_gcd = gcd(self.denom, self.numer)
        return (self.numer//the_gcd, self.denom//the_gcd)

    def __eq__(self, other):
        reduced_self = self.reduce_rational()
        reduced_other = other.reduce_rational()
        return reduced_self.numer == reduced_other.numer and \
            reduced_self.denom == reduced_other.denom

#test
a = Rational(2,7)
b = Rational(10, 31)

print(a + 5)
print(b - 10)
print(5 * a)
print(10/b)
```

```python
# Q7
class Odometer(object):
    def __init__(self, mileage, unit = 'miles') -> None:
        if mileage < 0 or (type(mileage) != int and type(mileage) != float):
            print('Milage entered cannot be negative or non-numeric, and its initialized to 0')
            mileage = 0

        if unit.lower() != 'miles' and unit.lower() != 'kilometers':
            print('Unit can only be miles or kilometers and its initialized to miles')
            unit = 'miles'

        self.mileage = mileage
        self.unit = unit.lower()

    def __str__(self) -> str:
        return f'Mileage:{self.mileage:.1f} {self.unit}'
    
    def __repr__(self):
        return self.__str__()

    def __add__(self, other):
        if type(other) == int or type(other) == float:
            self.mileage += other
            if self.mileage < 0:
                print(f'Mileage after operation = {self.mileage} < 0 so its reset to 0')
                self.mileage = 0
            return self
        else:
            print(f"'+' operand does not support {type(other)}")
            raise TypeError

    def __sub__(self, other):
        # equal to add negative of other
        return self.__add__(-other)

    def __radd__(self, other):
        return self.__add__(other)

    def __rsub__(self, other):
        if type(other) == int or type(other) == float:
            self.mileage = other - self.mileage
            if self.mileage < 0:
                print(f'Mileage after operation = {self.mileage:.1f} < 0 so its reset to 0')
                self.mileage = 0
            return self
        else:
            print(f"'-' operand does not support {type(other)}")
            raise TypeError



a = Odometer(500)
b = Odometer(200)
try:
    print(a + 100)
    print(50 - b)
    print(b - 10)
except TypeError:
    #error handling
    pass

try:
    print(a + b)
except TypeError:
    #error handling
    pass
```

```python
# Q8
class Clock(object):
    def __init__(self, hours = 0, minutes = 0, seconds = 0) -> None:
        if hours < 0 or hours > 24:
            hours, minutes, seconds = 0, 0, 0
        if minutes < 0 or minutes > 60:
            hours, minutes, seconds = 0, 0, 0
        if seconds < 0 or seconds > 60:
            hours, minutes, seconds = 0, 0, 0 
        
        self.hours = hours
        self.minutes = minutes
        self.seconds = seconds
    
    def __str__(self) -> str:
        return f'{self.hours:0>2}:{self.minutes:0>2}:{self.seconds:0>2}'

    def __repr__(self) -> str:
        return self.__str__()

    def __add__(self, other):
        if type(other) == int:
            return Clock((self.hours + other) % 24, self.minutes, self.seconds)
        elif type(other) == Clock:
            new_seconds = self.seconds + other.seconds
            minute_incre = new_seconds >= 60
            new_minute = self.minutes + other.minutes + minute_incre
            hour_incre = new_minute >= 60
            new_hours = self.hours + other.hours + hour_incre
            return Clock(new_hours % 24, new_minute % 60, new_seconds % 60)
        else:
            print(f'Addition Operation does not support {type(other)}')
            raise TypeError
    
    def __radd__(self, other):
        return self.__add__(other)

a = Clock(5, 59, 59)
b = Clock(18, 5, 0)
try:
    print(a + b)
    print(9 + b)
except TypeError:
    # error handle
    pass

try:
    print(a + 0.5)
except TypeError:
    # error handle
    pass
```
